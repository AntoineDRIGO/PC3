# Primero Hacer esta tarea: Integre los archivos que est√°n el folder comprimido (appendData.zip)



```python
import pandas as pd
#I start with the table "level1"
url = 'https://github.com/AntoineDRIGO/PC3/raw/main/level1.csv'

# I read csv precising "," is the way to get decimals
level1 = pd.read_csv(url, sep=';', decimal=',', encoding='ISO-8859-1')
level1
```

```python
#I rename the columns
level1 = level1.rename(columns={
    'HDI rank': 'HDI_rank_2021','Human Development Index (HDI)':'HDI','Life expectancy at birth':'Life_expectancy',
    'Expected years of schooling': 'School_level_expected','Mean years of schooling':'Mean_school_level',
    'Gross national income (GNI) per capita':'GNI','GNI per capita rank minus HDI rank': 'GNI_rank'})
level1
```

```python
print(level1.dtypes)
```

```python
# I delate the spaces in the GNI column to transform it in float
level1.GNI = level1.GNI.str.replace(' ', '')
level1.GNI = level1.GNI.astype(float)
type(level1.GNI[0])
```

# Segundo: Cuando haya integrado los archivos anteriores, haga un merge entre ese archivo, y el archivo countryCODES.xlsx.

```python
#now countrycodes
url = 'https://github.com/AntoineDRIGO/PC3/raw/main/countryCodes.csv' # Changed PC4 to PC3 in the url

# I read csv precising "," is the way to get decimals
countryCodes = pd.read_csv(url, sep=';', decimal=',', encoding='ISO-8859-1')
countryCodes
```

```python
#I join all the HDI database
HDI_db = pd.concat([level1, level2, level3, level4], ignore_index=True)
HDI_db
```

```python
#I delate specials caracters
HDI_db = HDI_db.applymap(lambda x: ''.join(c for c in str(x) if c.isalnum() or c.isspace()))

# I want to capitalise the entire Country column in HDI_db
HDI_db['Country'] = HDI_db['Country'].str.upper()
HDI_db
```

```python
# I merged th two tables
merged_df = pd.merge(HDI_db, countryCodes, on='Country', how='left')
merged_df
```

# Tercero: Con los archivos fallecidosCOVID.csv,  haga un reporte mensual de fallecidos. Solo debe mostrar por fila el mes (y a√±o) y cuanto fallecidos hubo en ese mes.

```python
#now countrycodes
url = 'https://github.com/AntoineDRIGO/PC3/raw/main/fallecidosCOVID.csv' # Changed PC4 to PC3 in the url

# I read csv precising "," is the way to get decimals
covid = pd.read_csv(url, sep=';', decimal=',', encoding='ISO-8859-1')
covid
```

```python
print(covid.dtypes)
```

```python
# I transform FECH_CORTE of covid on date

covid['FECHA_CORTE'] = pd.to_datetime(covid['FECHA_CORTE'], format='%Y%m%d')
covid['FECHA_FALLECIMIENTO'] = pd.to_datetime(covid['FECHA_FALLECIMIENTO'], format='%Y%m%d')
covid
```

```python
#  I create groups of same mounth of FECHA_FALLECIMIENTO and count the number of lines in each groups in mounth_death

monthly_deaths = covid.groupby(covid['FECHA_FALLECIMIENTO'].dt.to_period('M'))['FECHA_FALLECIMIENTO'].count()
print(monthly_deaths)
```

# Cuarto: Use el mismo archivo anterior. Esta vez haga un reporte de fallecidos por mes, pero el tipo de muerte debe estar en columnas separadas.

```python
# I delete all the special caracters in the columns strings

covid['CLASIFICACION_DEF'] = covid['CLASIFICACION_DEF'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
covid['DEPARTAMENTO'] = covid['DEPARTAMENTO'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
covid['PROVINCIA'] = covid['PROVINCIA'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
covid['DISTRITO'] = covid['DISTRITO'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)
covid
```

```python
# I make me sure that'CLASIFICACION_DEF' no have an infiry of values
print(covid['CLASIFICACION_DEF'].unique())
```

```python
# I create a new column Year_Month_Fallecimiento to group it after
covid['Year_Month_Fallecimiento'] = covid['FECHA_FALLECIMIENTO'].dt.to_period('M')
covid
```

```python
# I gorup by Year_Month_Fallecimiento and CLASIFICACION_DEF
monthly_report = covid.groupby(['Year_Month_Fallecimiento', 'CLASIFICACION_DEF']).size().unstack(fill_value=0)
monthly_report
```
